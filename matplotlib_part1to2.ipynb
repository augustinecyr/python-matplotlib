# Part 1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show()
```

plt.show() for non jupyter users

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.2])

axes1.plot(x,y)
axes1.set_title('Large')
axes2.plot(y,x)
axes2.set_title('Small')
```
